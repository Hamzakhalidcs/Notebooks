```python
# find the number of times when the speed is 4km/h
len(data[
    data["Wind Speed_km/h"] == 4
])
```

```python
# Rename the column of 'Weather' of the dataframe to 'Weather Condtion'
data.rename(columns= {'Weather': 'Weather Condition'}, inplace=True)
```

```python
data.head()
```

```python
# find the null values in the data 
data.isnull().sum()
```

```python
# finding the notnull values 
data.notnull().sum()
```

```python
# Find the visibily column mean value 
data.Visibility_km.mean()
```

```python
# to chk the standard deviation
data["Press_kPa"].std()
```

```python
# to chk the variance in column Rel Hum_%
data["Rel Hum_%"].var()
```

```python
# find all instances when the 'Snow' was recorded 
len(data[data["Weather Condition"] == "Snow"])
```

```python
# Finding the value in weather condtion in which snow is included 
data[
    data["Weather Condition"].str.contains("Sn[a-z]*", flags=re.I, regex=True)
    ]
```

```python
# Find all instance when windspeed is above 24 and visibility =25
data[
    (data["Wind Speed_km/h"]>24) &
    (data["Visibility_km"] == 25)
][["Wind Speed_km/h", "Visibility_km"]]
```

```python
# mean value aginst each weather conditon column 
data.groupby("Weather Condition").mean()
```

```python
# what is the min and max value against weather condition column 
```

```python
# for min 
data.groupby("Weather Condition").min()
```

```python
# for max
data.groupby("Weather Condition").max()
```

```python
# show all the records where weather conditoin is Fog
data[data["Weather Condition"].str.contains("fog", flags=re.I, regex=True)]
```

```python
# find record where weather is clear and visibility above 40 
data[
    (data["Weather Condition"] == "Clear") &
    (data["Visibility_km"] >40)
]
```

```python
# find record weather is clear and relative humidity is above 50 or visibility above 40
data[
    (data["Weather Condition"] == "Clear")&
    (data["Rel Hum_%"] >50) |
    (data["Visibility_km"]>40)
][["Weather Condition", "Rel Hum_%", "Visibility_km"]]
```
