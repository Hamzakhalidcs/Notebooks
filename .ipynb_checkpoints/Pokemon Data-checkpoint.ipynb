### Making Changes to Data

```python
del df["Total"]
```

```python
df.head(2)
```

```python
#  make new column by adding specfic columns (already in df, just for practice)
# Both lines will do the same but follow the second line for best practices

df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
df["Total"] = df.iloc[:, 4:10].sum(axis=1)
```

```python
# Both lines will do the same job

df.drop(columns=['Total'], inplace=True)
df = df.drop(columns=['Total'])
```

```python
# Get Total column before HP column

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-2]] +[cols[-1]] + cols[4:11]]
```

```python
# df.iloc[: , -2]
df.head()
```

```python
# saving this dataframe in modofied file

df.to_csv("modified.csv", index=False)
df.to_excel("modified.xlsx", index=False)
df.to_csv("modified.txt", index=False, sep="\t ")
```

```python
import re  # Regulare Expression

df.loc[
    df["Type 1"].str.contains("grass|fire", flags=re.I, regex=True)  #re.I means ignore case sensitive
]
```

## Conditional Changes

```python
# changes fire to flamer i.e if "Type 1" is = "Fire" change "Type 1" value to "Flammer"

df.loc[
    df["Type 1"] == "Fire", "Type 1"
] = "Flamer"
```

```python
#  if type 1 = fire set legendary column to true 
df.loc[
    df["Type 1"] == "fire", "Legendary"
] = True
```

```python
df[
    df["Type 1"] == "fire"
][["Type 1", "Legendary"]].head(2)
```

```python
df.loc[
    (df["Type 1"] == "fire") &
    (df["Name"].str.contains("Mega")) & 
    (df["HP"]>500), "Legendary"
] = True
```

```python
df.loc[
    (df["Type 1"].str.contains("fire")) & 
    (df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)) &
    (df["HP"]>500), ["Generation", "Legendary"]
] = [0 , False]
```

```python
df.loc[df["Total"]>500, ["Generation", "Legenedary"]] = [0, False]
```

```python
df.loc[
    df["Total"]>500, ["Generation", "Legendary"]
] = [0, False]
```

```python
df[
    df["Total"]>500
][["Total", "Generation", "Legendary"]].head()
```

### Aggregate Statistics (Groupby)

```python
# To see the Averages

df.groupby(["Type 1"]).mean()
```

```python
df["Type 1"].value_counts()
```

```python
# sort values base on value in Type 1 like Defense
 
df.groupby(["Type 1"]).mean().sort_values('Defense', ascending=False)
```

```python
# using sum fucntion 
df.groupby(['Type 1']).sum()
```

```python
# Added the new column to DataFrame
df["count"] = 1
```

```python
df.groupby(["Type 1"]).count()["count"]
```

```python
# groupby multiple columns
df.groupby(["Type 1", "Type 2"]).count()["count"]
```
