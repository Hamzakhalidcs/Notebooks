```python
import pandas as pd

# Load the data
df = pd.read_csv("Diwali Sales Data.csv", encoding='latin1')

# --- ETL: Clean the dataset ---

# Drop unnecessary columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)

# Remove rows with missing amount
df = df[df['Amount'].notnull()]

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Clean special characters in 'state'
df['state'] = df['state'].str.replace('\xa0', ' ')

# --- EDA: Extract insights ---

# Shape of the data
print("Shape:", df.shape)

# Null values
print("\nNull values:\n", df.isnull().sum())

# Data types
print("\nData types:\n", df.dtypes)

# Unique values per column
print("\nUnique values:\n", df.nunique())

# Top 5 states by total sales
print("\nTop 5 States by Sales:\n", df.groupby("state")["amount"].sum().sort_values(ascending=False).head())

# Top 5 product categories by sales
print("\nTop 5 Product Categories by Sales:\n", df.groupby("product_category")["amount"].sum().sort_values(ascending=False).head())

# Top 5 users by total spend
print("\nTop 5 Users by Spending:\n", df.groupby("user_id")["amount"].sum().sort_values(ascending=False).head())
```
