```python
import pandas as pd 
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
# use map function to change the value in the column 
train['sex_num'] = train.Sex.map({'female' : 0, 'male' : 1})
```

```python
train.loc[0:4, ['Sex', 'sex_num']]
```

```python
# Add new column for the length of Name column record len
train["Name_length"] = train.Name.apply(len)
```

```python
train[["Name", "Name_length"]].head()
```

```python
train.loc[
    train["Name_length"] == 24, "Name_length"
] = 23
```

```python
train[
    train["Name_length"] == 23
][["Name", "Name_length"]].head(2)
```

```python
train.loc[
    train["Name_length"] == 24, "Name_length"
] = 23
```

```python
train[
    train["Name_length"]==23
].head()
```

```python
train["sex_num2"]  = train["Sex"].map({'female':0, 'male':1})
```

```python
train["Name_length_2"] = train["Name"].apply(len)
```

```python
train[["Name", "Sex", "sex_num", "sex_num2", "Name_length", "Name_length_2"]].head()
```

```python
train['Fare'].head()
```

```python
# To round the figure
import numpy as np 
train["Fare_ceil"] = train.Fare.apply(np.ceil)
```

```python
train[['Fare', 'Fare_ceil']].head()
```

```python
# split on , in Names column
train.Name.str.split(',').head()
```

```python
train["Name_list"] = train["Name"].str.split(',')
```

```python
train[["Name", "Name_list", "first_name", "first_name_fun", "last_name", "first_name_lambda"]].head()
```

```python
train["first_name"] = train["Name_list"].str[0]
```

```python
train["last_name"] = train["Name_list"].str[1]
```

```python
train["Name"].str.split(',')
```

```python
# same work can do using function
def get_element(my_list, position):
    return my_list[position]
```

```python
train["first_name_fun"] = train["Name"].str.split(',').apply(get_element, position=0)
```

```python
# same work using lambda function
train["first_name_lambda"] = train.Name.str.split(',').apply(lambda x:x[0]).head()
```
